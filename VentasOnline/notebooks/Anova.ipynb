Librerías necesarias

```python
# Manipulación de datos
import pandas as pd

# Estadística
from scipy import stats
```

Cargar el dataset

```python
# Cargar el dataset
df = pd.read_csv(
    "C:/Users/analu/Downloads/Curso_python/VentasOnline/data/ventas_online.csv"
)

# Convertir fecha a datetime (buena práctica)
df["fecha"] = pd.to_datetime(df["fecha"])
```

Exploración previa

```python
# Promedio de satisfacción por categoría
df.groupby("categoria")["satisfaccion"].mean()
```

Revisar tamaño de muestra por grupo

```python
# Cantidad de observaciones por categoría
df["categoria"].value_counts()
```

Formular las hipótesis


Hipótesis nula (H₀)
La satisfacción promedio es igual en todas las categorías

Hipótesis alternativa (H₁)
Al menos una categoría tiene una satisfacción promedio diferente


Preparar los datos para ANOVA

```python
# Extraer la satisfacción de cada categoría
satisf_elec = df[df["categoria"] == "Electrónica"]["satisfaccion"]
satisf_hogar = df[df["categoria"] == "Hogar"]["satisfaccion"]
satisf_ropa = df[df["categoria"] == "Ropa"]["satisfaccion"]
satisf_deportes = df[df["categoria"] == "Deportes"]["satisfaccion"]
satisf_libros = df[df["categoria"] == "Libros"]["satisfaccion"]
```

```python
# Aplicar ANOVA de un factor
f_stat, p_value = stats.f_oneway(
    satisf_elec,
    satisf_hogar,
    satisf_ropa,
    satisf_deportes,
    satisf_libros
)

f_stat, p_value
```

Interpretación de resultados

```python
# Nivel de significancia
alpha = 0.05

if p_value < alpha:
    print("❌ Rechazamos H0: existen diferencias significativas entre categorías")
else:
    print("✅ No rechazamos H0: no hay diferencias significativas entre categorías")
```
